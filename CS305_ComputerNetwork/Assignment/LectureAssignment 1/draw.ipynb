```python
import numpy as np
import matplotlib.pyplot as plt

# 创建一个包含自变量 N 的范围
N = np.linspace(1, 1000, 1000)

# 计算函数值
Dcs = np.maximum(N * 20e9 / 15e6, 20e9 / 4e6)
D1 = np.maximum(5000, N * 20e9/(15e6 + N * 1e5))
D2 = np.maximum(5000, N * 20e9/(15e6 + N * 6e5))
D3 = np.maximum(5000, N * 20e9/(15e6 + N * 4e6))

# 创建图形
plt.plot(N, Dcs, label='client-service')
plt.plot(N, D1, label='P2P u=100K')
plt.plot(N, D2, label='P2P u=600K')
plt.plot(N, D3, label='P2P u=4M')

# 添加标签和标题
plt.xlabel('N')
plt.ylabel('Time')

# 显示图例
plt.legend()

# 显示图形
plt.grid(True)
plt.show()
```
