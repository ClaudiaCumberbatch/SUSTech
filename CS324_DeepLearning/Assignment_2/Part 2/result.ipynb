```python
# 提取训练和测试的损失和准确率
epochs = []
train_losses = []
test_losses = []
train_accuracies = []
test_accuracies = []

epoch_pattern = re.compile(r"Epoch (\d+), Train Loss: ([\d.]+), Train Accuracy: ([\d.]+)%")
test_pattern = re.compile(r"Epoch \d+, Test Loss: ([\d.]+), Test Accuracy: ([\d.]+)%")

for match in epoch_pattern.finditer(log_data):
    epoch, train_loss, train_acc = match.groups()
    epochs.append(int(epoch))
    train_losses.append(float(train_loss))
    train_accuracies.append(float(train_acc))

for match in test_pattern.finditer(log_data):
    test_loss, test_acc = match.groups()
    test_losses.append(float(test_loss))
    test_accuracies.append(float(test_acc))

# 训练和测试损失图
plt.figure(figsize=(8, 5))
plt.plot(epochs, train_losses, label='Train Loss')
plt.plot(epochs, test_losses, label='Test Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Train and Test Loss')
plt.legend()
plt.savefig('images/cnn_loss.pdf')
plt.show()

# 训练和测试准确率图
plt.figure(figsize=(8, 5))
plt.plot(epochs, train_accuracies, label='Train Accuracy')
plt.plot(epochs, test_accuracies, label='Test Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy (%)')
plt.title('Train and Test Accuracy')
plt.legend()
plt.savefig('images/cnn_accuracy.pdf')
plt.show()
```
